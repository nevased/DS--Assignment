# Applying naive bayes for classification

```python
# Preparing a Multinomial naive bayes model on training data set 

from sklearn.naive_bayes import MultinomialNB as MB
# Multinomial Naive Bayes

# Model Train
classifier_mb = MB()

classifier_mb.fit(x_train,y_train)

# Model Test

# Model Accuracy on train set
train_pred_m = classifier_mb.predict(x_train)
accuracy_train_m = np.mean(train_pred_m==y_train) 

# Model Accuracy on test set
test_pred_m = classifier_mb.predict(x_train)
accuracy_test_m = np.mean(test_pred_m==y_train)
```

```python
accuracy_train_m
```

```python
accuracy_test_m
```

# After applying multinomial naive bayes we get same accuracy for train and test dataset i.e 0.77

```python
# Preparing a Gaussian naive bayes model on training data set 

from sklearn.naive_bayes import GaussianNB as GB
# Gaussian Naive Bayes

# Model Train
classifier_gb = GB()

classifier_gb.fit(x_train,y_train)

# Model Test

# Model Accuracy on train set
train_pred_g = classifier_gb.predict(x_train)
accuracy_train_g = np.mean(train_pred_g==y_train) 

# Model Accuracy on test set
test_pred_g = classifier_gb.predict(x_train)
accuracy_test_g = np.mean(test_pred_g==y_train)
```

```python
accuracy_train_g
```

```python
accuracy_test_g
```

# After applying Gaussian naive bayes we get same accuracy for train and test dataset 0.78

```python
# Preparing a Complement naive bayes model on training data set 

from sklearn.naive_bayes import ComplementNB as CB
# Complement Naive Bayes

# Model Train
classifier_cb = CB()

classifier_cb.fit(x_train,y_train)

# Model Test

# Model Accuracy on train set
train_pred_cb = classifier_cb.predict(x_train)
accuracy_train_cb = np.mean(train_pred_cb==y_train) 

# Model Accuracy on test set
test_pred_cb = classifier_cb.predict(x_train)
accuracy_test_cb = np.mean(test_pred_cb==y_train)
```

```python
accuracy_train_cb
```

```python
accuracy_test_cb
```

# After applying Complement naive bayes we get same accuracy for train and test dataset i.e 0.77

```python
# Preparing a Categorical naive bayes model on training data set 

from sklearn.naive_bayes import CategoricalNB as CNB
# Categorical Naive Bayes

# Model Train
classifier_cnb = CNB()

classifier_cnb.fit(x_train,y_train)

# Model Test

# Model Accuracy on train set
train_pred_cnb = classifier_cnb.predict(x_train)
accuracy_train_cnb = np.mean(train_pred_cnb==y_train) 

# Model Accuracy on test set
test_pred_cnb = classifier_cnb.predict(x_train)
accuracy_test_cnb = np.mean(test_pred_cnb==y_train)
```

```python
accuracy_train_cnb
```

```python
accuracy_test_cnb
```

# After applying Categorical naive bayes we get same accuracy for train and test dataset.i.e 0.81
